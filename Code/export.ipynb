Ploting Days of month Vs quantity

```python
df['Month'] = df['SailDate'].dt.month
df['Day'] = df['SailDate'].dt.day
```

```python
daily_quantity = df.groupby(['Month', 'Day'])['QTY.(KGS)'].sum().reset_index()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming daily_quantity is defined and contains your data

# Create subplots
fig, axs = plt.subplots(3, 4, figsize=(20, 15))

# Define month names
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Loop through each month
for month in range(1, 13):
    # Filter data for the current month
    month_data = daily_quantity[daily_quantity['Month'] == month]
    
    # Get the corresponding subplot
    ax = axs[(month - 1) // 4, (month - 1) % 4]
    
    # Plot data
    ax.plot(month_data['Day'], month_data['QTY.(KGS)'])
    
    # Set title with month name
    ax.set_title(months[month - 1])
    
    # Set axis labels
    ax.set_xlabel('Day of Month')
    ax.set_ylabel('Quantity')

# Adjust layout
plt.tight_layout()
plt.show()
```

Ploting the variation of sales in each quater in Detail

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import numpy as np

# Assuming you have a DataFrame named df with the relevant data

# Convert 'Sail Date.' column to datetime if not already in datetime format
df['SailDate'] = pd.to_datetime(df['SailDate'])

unique_quarters = df['Quarter'].unique()

# Create a color map for quarters
colors = cm.rainbow_r(np.linspace(0, 1, len(unique_quarters)))

# Plot each quarter's export with different color
plt.figure(figsize=(10, 6))
for i, quarter in enumerate(unique_quarters):
    quarter_data = df[df['Quarter'] == quarter]
    plt.plot(quarter_data['SailDate'], quarter_data['QTY.(KGS)'], marker='o', label=f'Q{quarter.quarter} {quarter.year}', color=colors[i])

plt.title('Export Quantity by Quarter')
plt.xlabel('Date')
plt.ylabel('Total Quantity (KGS)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

ACF

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Convert 'SailDate' to datetime
df['SailDate'] = pd.to_datetime(df['SailDate'])

# Check if the specified column exists in the DataFrame
if 'SailDate' in df.columns:
    # Resample data to compute monthly sales quantity
    df['Month'] = df['SailDate'].dt.to_period('M')  # Add a new column 'Month' with monthly periods
    monthly_quantity = df.groupby('Month')['QTY.(KGS)'].sum()  # Group by month and sum quantities
    
    # Plot ACF
    plt.figure(figsize=(10, 6))
    plot_acf(monthly_quantity, lags=11)  # Assuming maximum lag of 11 months
    plt.xlabel('Lag (Months)')
    plt.ylabel('Autocorrelation')
    plt.title('Autocorrelation Function (ACF)')
    plt.xticks(range(12))  # Assuming maximum lag of 11 months
    plt.grid(True)
    plt.show()
else:
    print("Error: 'SailDate' column not found in the DataFrame.")
```
